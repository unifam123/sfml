## DecisionTreeClassifier


```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
dtc_test = DecisionTreeClassifier(random_state=42)
params = {'max_depth': [i for i in range(1, 100, 2)]}
tree_test = GridSearchCV(dtc_test, params, scoring='roc_auc', n_jobs=-1)
```

```python
%time tree_test.fit(train, y)
```

```python
tree_results = pd.DataFrame(tree_test.cv_results_)
```

```python
tree_results.sort_values('mean_test_score').tail()
```

```python
tree_test.best_params_
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
fig, ax = plt.subplots(figsize=(20, 10))

plt.rc('font', **{'size'   : 20})

plt.plot(tree_results['param_max_depth'], tree_results['mean_test_score'], 
         label='roc_auc: '+str(tree_test.best_score_)+u'\nЛучшая глубина: '+str(tree_test.best_params_['max_depth']))

plt.legend()
plt.grid()
plt.xlabel(u'Глубина дерева')
plt.ylabel(u'ROC_AUC_SCORE')
```

# KNN

```python
knn_test = KNeighborsClassifier()
knn_params = {'n_neighbors': [i for i in range(1, 50, 4)]}
knn_grid = RandomizedSearchCV(knn_test, knn_params, scoring='roc_auc', n_jobs=4, cv=2)
```

```python
len(knn_params['n_neighbors'])
```

```python
#урежим пололвину признаков. Иначе будет до второго пришествия пересчитывать
dropped_train = train.drop(train.columns[[i for i in range(0, len(train.columns), 2)]], axis=1)
dropped_train = dropped_train.drop(dropped_train.columns[[i for i in range(0, len(dropped_train.columns), 2)]], axis=1)
```

```python
dropped_train.head()
```

```python
%time knn_grid.fit(dropped_train, y)
```

```python
knn_results = pd.DataFrame(knn_grid.cv_results_).sort_values('param_n_neighbors')
```

```python
fig, ax = plt.subplots(figsize=(20, 10))


plt.plot(knn_results['param_n_neighbors'], knn_results['mean_test_score'], 
         label='roc_auc: '+str(knn_grid.best_score_)+u'\nЛучшее количество соседей: '+str(knn_grid.best_params_['n_neighbors']))

plt.legend()
plt.grid()
plt.xlabel(u'Количество соседей')
plt.ylabel(u'ROC_AUC_SCORE')
```

# LogisticRegression

```python
lr_test = LogisticRegression()
lr_params = {'penalty': ('l1', 'l2'), 'C': [100., 1., 0.1]}
lr_grid = GridSearchCV(lr_test, lr_params, scoring='roc_auc', n_jobs=4, cv=2)
```

```python
%time lr_grid.fit(train, y)
```

```python
lr_results = pd.DataFrame(lr_grid.cv_results_)
```

```python
lr_results.head()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from matplotlib import cm
import matplotlib as mpl



fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(111, projection='3d')

X=np.array(lr_results['param_penalty'])
Y=np.array(lr_results['param_C'])
Z=np.array(lr_results['mean_test_score'])

X[X=='l1'] = 1
X[X=='l2'] = 2

cm = cm.get_cmap('RdYlBu')

plt.xticks([1, 2], ["l1", "l2"])

surf = ax.scatter(X, Y, Z, c=Z, s=400, cmap=cm, label='roc_auc: '+str(lr_grid.best_score_)+u'\nЛучший регуляризатор: '+str(lr_grid.best_params_['penalty'])
                                                                         +'\n Лучший C: '+ str(lr_grid.best_params_['C']))

ax.legend()
ax.set_zlim(0.6, 0.9)
ax.set_xlabel(u"Регуляризатор")
ax.set_ylabel(u"Гиперпараметр C")
ax.set_zlabel(u"ROC_AUC")

plt.show()
```
